```python
import bs4
import pandas as pd
from selenium import webdriver
import time
```

* 통화 : currency
* 현찰 사실 때 환율 : buy
* 현찰 사실 때 Spread : buy_spread
* 현찰 파실 때 환율 : sell
* 현찰 파실 때 Spread : sell_spread
* 송금 보낼 때 : send
* 송금 받을 때 : receive
* T/C 사실 때 : buy_TC
* 외화수표파실 때 : sell_check
* 매매기준율 : stdexrate
* 환가료율 : exchange_rate
* 미화환산율 : dollarexrate

```python
column_list = ["currency","buy", "buy_spread","sell", "sell_spread","send","receive","buy_TC","sell_TC","stdexrate","exchange_rate","dollarexrate",'Date']
```

```python
def hana_exchange(date_):
    
    currency = [];
    buy = []
    buy_spread = []
    sell = []
    sell_spread = []
    send = []
    receive = []
    buy_TC = []
    sell_TC = []
    stdexrate = []
    exchange_rate = []
    dollarexrate = []
    Date = []

    options = webdriver.ChromeOptions()
    # options.add_argument('headless')
    driver = webdriver.Chrome('D:\Chromdriver\chromedriver.exe', options=options)
    driver.get('https://www.hanabank.com/cont/mall/mall15/mall1501/index.jsp')
    time.sleep(2)
    driver.switch_to.frame('bankIframe')
    driver.find_element_by_name('tmpInqStrDt').click()
    driver.find_element_by_name('tmpInqStrDt').clear()
    driver.find_element_by_name('tmpInqStrDt').send_keys(date_)
    driver.find_element_by_xpath('//a[@class="btnDefault bg"]').click()
    driver.find_element_by_id('searchContentDiv')
    bsObj = bs4.BeautifulSoup(driver.page_source, "html.parser")
    bsObj_list = bsObj.find('table', {"class": "tblBasic leftNone"}).tbody.find_all('tr')

    for exchange_info in bsObj_list:
        currency.append(exchange_info.find('td', {"class": "tc"}).get_text().strip())
        buy.append(exchange_info.find_all('td', {"class": "txtAr"})[0].get_text())
        buy_spread.append(exchange_info.find_all('td', {"class": "txtAr"})[1].get_text())
        sell.append(exchange_info.find_all('td', {"class": "txtAr"})[2].get_text())
        sell_spread.append(exchange_info.find_all('td', {"class": "txtAr"})[3].get_text())
        send.append(exchange_info.find_all('td', {"class": "txtAr"})[4].get_text())
        receive.append(exchange_info.find_all('td', {"class": "txtAr"})[5].get_text())
        buy_TC.append(exchange_info.find_all('td', {"class": "txtAr"})[6].get_text())
        sell_TC.append(exchange_info.find_all('td', {"class": "txtAr"})[7].get_text())
        stdexrate.append(exchange_info.find_all('td', {"class": "txtAr"})[8].get_text())
        exchange_rate.append(exchange_info.find_all('td', {"class": "txtAr"})[9].get_text())
        dollarexrate.append(exchange_info.find_all('td', {"class": "txtAr"})[10].get_text())
        Date.append(date_)

    df2 = pd.DataFrame(list(zip(currency, buy, buy_spread, sell, sell_spread, send, receive, buy_TC, sell_TC, stdexrate, exchange_rate, dollarexrate, Date)), columns=column_list)

    return df2
```

```python
hana_exchange('2021-03-01')
```
